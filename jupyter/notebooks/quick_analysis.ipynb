```python
devices_df['time'] = pd.to_datetime(devices_df['time'], unit='s')# Set the index of the DataFrame to the 'time' column
devices_df.set_index('time', inplace=True)
print(devices_df.columns)

# Initialize empty lists to store the results
max_temps_per_hour = []
data_points_per_hour = []

# Iterate over each device_id
for device_id, device_df in devices_df.groupby('device_id'):
    # Get the maximum temperature per hour for the current device_iddevices_df['time'] = devices_df['time'].apply(lambda x: datetime.datetime.utcfromtimestamp(x))

    max_temp = device_df['temperature'].resample('H').max()
    max_temps_per_hour.append(max_temp)

    # Get the number of data points per hour for the current device_id
    data_points = device_df['temperature'].resample('H').count()
    data_points_per_hour.append(data_points)

# Combine the results into DataFrames
max_temps_per_hour_df = pd.concat(max_temps_per_hour, axis=1)
max_temps_per_hour_df.columns = devices_df['device_id'].unique()
data_points_per_hour_df = pd.concat(data_points_per_hour, axis=1)
data_points_per_hour_df.columns = devices_df['device_id'].unique()

# Pivot the DataFrames to have separate lines for each device_id
max_temps_per_hour_pivoted = max_temps_per_hour_df.stack().reset_index().rename(columns={'level_0': 'time', 'level_1': 'device_id', 0: 'max_temperature'})
data_points_per_hour_pivoted = data_points_per_hour_df.stack().reset_index().rename(columns={'level_0': 'time', 'level_1': 'device_id', 0: 'data_points'})

# Set the line width to 1000 characters
pd.set_option('display.width', 200)

# Display the results
print("Maximum temperatures per hour for each device:")
print(max_temps_per_hour_pivoted)
print("\nNumber of data points per hour for each device:")
print(data_points_per_hour_pivoted)
```

```python
# Initialize empty lists to store the results
max_temps_per_interval = []
data_points_per_interval = []

# Iterate over each device_id
for device_id, device_df in devices_df.groupby('device_id'):
    # Get the maximum temperature per 10-minute interval for the current device_id
    max_temp = device_df['temperature'].resample('1T').max()
    max_temps_per_interval.append(max_temp)

    # Get the number of data points per 10-minute interval for the current device_id
    data_points = device_df['temperature'].resample('1T').count()
    data_points_per_interval.append(data_points)

# Combine the results into DataFrames
max_temps_per_interval_df = pd.concat(max_temps_per_interval, axis=1)
max_temps_per_interval_df.columns = devices_df['device_id'].unique()
data_points_per_interval_df = pd.concat(data_points_per_interval, axis=1)
data_points_per_interval_df.columns = devices_df['device_id'].unique()

# Pivot the DataFrames to have separate lines for each device_id
max_temps_per_interval_pivoted = max_temps_per_interval_df.stack().reset_index().rename(columns={'level_0': 'time', 'level_1': 'device_id', 0: 'max_temperature'})
data_points_per_interval_pivoted = data_points_per_interval_df.stack().reset_index().rename(columns={'level_0': 'time', 'level_1': 'device_id', 0: 'data_points'})

# Sort the results by device_id and time
max_temps_per_interval_sorted = max_temps_per_interval_pivoted.sort_values(by=['device_id', 'time'])
data_points_per_interval_sorted = data_points_per_interval_pivoted.sort_values(by=['device_id', 'time'])

# Set the line width to 1000 characters and display all rows
pd.set_option('display.width', 200)
pd.set_option('display.max_rows', None)

# Display the sorted results
print("Maximum temperatures per 10-minute interval for each device:")
print(max_temps_per_interval_sorted)
print("\nNumber of data points per 10-minute interval for each device:")
print(data_points_per_interval_sorted)
```

```python
# Sort the DataFrame by device_id and time
devices_df.sort_values(by=['device_id', 'time'], inplace=True)
print(device_df.columns)
print(devices_df.columns)
```

```python

# Initialize an empty DataFrame to store the results
distances_df = pd.DataFrame(columns=['device_id', 'time', 'distance'])

# Iterate over each device_id
for device_id, device_df in devices_df.groupby('device_id'):
    # Calculate the distance between consecutive data points
    distances = device_df.apply(
        lambda row: great_circle(
            (row['latitude'], row['longitude']),
            (device_df.shift(1).loc[row.name]['latitude'], device_df.shift(1).loc[row.name]['longitude'])
        ).kilometers if row.name > device_df.index[0] else None,
        axis=1
    )
    
    # Add the results to the distances_df DataFrame
    temp_df = pd.DataFrame({
        'device_id': [device_id] * len(device_df),
        'time': device_df.index,
        'distance': distances
    })
    
    #distances_df = distances_df.append(temp_df, ignore_index=True)
    distances_df = pd.concat([distances_df, temp_df], ignore_index=True)

# Display the results
print(distances_df)
```

```python
# Reset the index to make 'time' a regular column
distances_df.reset_index(inplace=True)
```

```python
# Create an empty DataFrame to store the results
summary_df = pd.DataFrame(columns=['device_id', 'hour', 'distance', 'count'])


# Calculate the summary of distances for each device and hour
for device_id, device_group in distances_df.groupby('device_id'):
    for hour, hour_group in device_group.groupby(pd.Grouper(key='time', freq='H')):
        distance_sum = hour_group['distance'].sum()
        count = hour_group['distance'].count()
        temp_df = pd.DataFrame({
            'device_id': [device_id],
            'hour': [hour],
            'distance': [distance_sum],
            'count': [count]
        })
        summary_df = pd.concat([summary_df, temp_df], ignore_index=True)
        
# Format the distance column to show distances in kilometers with 2 decimal places
summary_df['distance'] = summary_df['distance'].apply(lambda x: '{:.2f} km'.format(x))

# Display the results
print(summary_df)
```
